Create a metaclass named MyMeta that inherits from type. The metaclass should have a method named get_attrs that returns a list of all the attributes of the class it creates.

Example usage:

```Python
class MyClass(metaclass=MyMeta):
    foo = 1
    bar = "hello"

attrs = MyClass.get_attrs()
print(attrs) # ['__module__', '__qualname__', 'foo', 'bar']
```
